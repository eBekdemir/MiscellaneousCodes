```python
import pandas as pd
```

```python
top250 = pd.read_parquet('top250.parquet')
suggested_by250 = pd.read_parquet('suggested_by250.parquet')
suggested_level2 = pd.read_parquet('suggested_level2.parquet')
suggested_level3 = pd.read_parquet('suggested_level3.parquet')
suggested_level4 = pd.read_parquet('suggested_level4.parquet')
```

```python
all_movies = pd.concat([top250, suggested_by250, suggested_level2, suggested_level3, suggested_level4])
all_movies = all_movies.drop_duplicates(subset='id')
```

```python
all_movies = all_movies.drop(columns=['suggestions', 'released'])
```
